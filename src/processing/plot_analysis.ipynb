```python
import pickle
import matplotlib.pyplot as plt

with open("vae_training_history.pkl", "rb") as fp:
    history = pickle.load(fp)

train_recon = history["train_recon"]
train_kl    = history["train_kl"]
train_total = history["train_total"]
train_acc   = history["train_acc"]
train_f1    = history["train_f1"]
val_recon   = history["val_recon"]
val_kl      = history["val_kl"]
val_total   = history["val_total"]
val_acc     = history["val_acc"]
val_f1      = history["val_f1"]

num_epochs = len(train_total)
epochs = list(range(1, num_epochs + 1))

plt.figure()
plt.plot(epochs, train_total, label="Train Total Loss")
plt.plot(epochs, val_total,   label="Val   Total Loss")
plt.xlabel("Epoch")
plt.ylabel("Total Loss")
plt.title("VAE Total Loss over Epochs")
plt.legend()
plt.grid(True)

plt.figure()
plt.plot(epochs, train_recon, label="Train Recon Loss")
plt.plot(epochs, val_recon,   label="Val   Recon Loss")
plt.xlabel("Epoch")
plt.ylabel("Reconstruction Loss")
plt.title("VAE Reconstruction Loss over Epochs")
plt.legend()
plt.grid(True)

plt.figure()
plt.plot(epochs, train_kl, label="Train KL Divergence")
plt.plot(epochs, val_kl,   label="Val   KL Divergence")
plt.xlabel("Epoch")
plt.ylabel("KL Divergence")
plt.title("VAE KL Divergence over Epochs")
plt.legend()
plt.grid(True)

plt.figure()
plt.plot(epochs, train_acc, label="Train Accuracy")
plt.plot(epochs, val_acc,   label="Val   Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Token-level Accuracy")
plt.title("Token‑level Accuracy over Epochs")
plt.legend()
plt.grid(True)

plt.figure()
plt.plot(epochs, train_f1, label="Train F₁ (macro)")
plt.plot(epochs, val_f1,   label="Val   F₁ (macro)")
plt.xlabel("Epoch")
plt.ylabel("F₁ Score")
plt.title("F₁ Score (macro) over Epochs")
plt.legend()
plt.grid(True)

plt.show()
```
